### Python program to draw a line in a diagram from position (0,0) to position (8,320)  

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 8])
ypoints = np.array([0, 320])

plt.plot(xpoints, ypoints)
plt.show()
```

---  
* <code>import matplotlib.pyplot as plt</code>: Importing matplotlib library with the name plt
* <code>import numpy as np</code>: Importing numpy library with the name np
* <code>xpoints = np.array([0, 8])</code>: Creating a numpy array having start and end points 0 and 8
* <code>ypoints = np.array([0, 320])</code>: Creating a numpy array having start and end points 0 and 320
* <code>plt.plot(xpoints, ypoints)</code>: Creating a line chart with respective x and y points given
* <code>plt.show()</code>: Displaying the plot
